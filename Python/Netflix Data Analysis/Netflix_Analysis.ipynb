## missing values 

```python
data.isnull().sum()
```

```python
#fill missing values
data['director'].fillna('Unknown', inplace=True)
data['cast'].fillna('Unknown', inplace=True)
data['country'].fillna('Unknown', inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data.duplicated().sum()
```

## Movie vs TV Show Analysis

```python
plt.figure(figsize=(7,5))
sns.countplot(x='type', data=data, palette=['#E50914', '#221F1F'])
plt.title('Comparison of TV Shows and Movies')
plt.show()

# Pie Chart code
plt.figure(figsize=(7,5))
plt.pie(data['type'].value_counts(), labels=data['type'].value_counts().index, 
        explode=(0.1, 0), autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Distribution of Content')
plt.show()
```

## Top 10 Countries

```python
top_countries = data['country'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='coolwarm')
plt.xticks(rotation=45)
plt.title('Top 10 Countries with Most Content')
plt.show()
```

## Month-Wise Content Addition 

```python
#extra spaces  remove removed
data['date_added'] = data['date_added'].str.strip()

#  date  conversion
data['date_added'] = pd.to_datetime(data['date_added'])

# Month find
data['month_added'] = data['date_added'].dt.month_name()

# Visualization 
plt.figure(figsize=(12,6))
sns.countplot(x='month_added', data=data, order=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'], palette='viridis')
plt.title('Content Added Month-Wise')
plt.xticks(rotation=45)
plt.show()
```

## Top Directors

```python
# 'Unknown'  filtered then directors findout
directors = data[data['director'] != 'Unknown']['director'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=directors.values, y=directors.index, palette='magma')
plt.title('Top 10 Title Hosting Directors')
plt.show()
```

### Word Cloud

```python
from wordcloud import WordCloud

plt.figure(figsize=(12,8))
wordcloud = WordCloud(background_color='black', width=1920, height=1080).generate(" ".join(data['title']))
plt.imshow(wordcloud)
plt.axis('off')
plt.title('Netflix Titles Word Cloud')
plt.show()
```

## Ratings Analysis

```python
plt.figure(figsize=(12,6))
sns.countplot(x='rating', data=data, order=data['rating'].value_counts().index, palette='Set2')
plt.title('Distribution of Ratings')
plt.show()
```

## Release Year Trends

```python
plt.figure(figsize=(12,6))
sns.histplot(data['release_year'], bins=30, kde=True, color='red')
plt.title('Content Release Year Distribution')
plt.show()
```
